```python
import numpy as np
import matplotlib.pyplot as plt
from pds4_tools import pds4_read

# Path to the XML file
xml_file = r"F:\ISRO_BAH\archive_files\ch2_ohr_ncp_20240425T140\data\calibrated\20240425\ch2_ohr_ncp_20240425T1406019344_d_img_d18.xml"

# Read the PDS4 file
data = pds4_read(xml_file)

# Extract image data
image_data = data[0].data

# Display the image using matplotlib
plt.imshow(image_data, cmap='gray')
plt.title('OHRC Image')
plt.show()

# Save the image using matplotlib
output_image_path = 'output_image.png'
plt.imsave(output_image_path, image_data, cmap='gray')

print(f"Image saved as {output_image_path}")
```
