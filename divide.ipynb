```python
from PIL import Image
import os

Image.MAX_IMAGE_PIXELS = None

def divide_image(image_path, output_folder, segment_width, segment_height):
    # Open the input image
    image = Image.open(image_path)
    image_width, image_height = image.size

    # Create the output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop to divide the image
    segment_number = 0
    for top in range(0, image_height, segment_height):
        for left in range(0, image_width, segment_width):
            # Calculate the bounds of the segment
            right = min(left + segment_width, image_width)
            bottom = min(top + segment_height, image_height)

            # Crop the image and save the segment
            segment = image.crop((left, top, right, bottom))
            segment_filename = os.path.join(output_folder, f"segment_{segment_number}.png")
            segment.save(segment_filename)
            segment_number += 1

# Define the path to the input image and the output folder
image_path = r"C:\Users\adith\Downloads\Lunar_LRO_LROC-WAC_Mosaic_global_100m_June2013.tif"
output_folder = 'output_segments_mosaic'

# Define the size of each segment
segment_width = 1000
segment_height = 1000

# Divide the image
divide_image(image_path, output_folder, segment_width, segment_height)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
def get_image_resolution(image_path):
    try:
        with Image.open(image_path) as img:
            plt.imshow(img)
            width, height = img.size
            return width, height
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
image_path = r"C:\Users\adith\Downloads\Lunar_LRO_LROC-WAC_Mosaic_global_100m_June2013.tif"
resolution = get_image_resolution(image_path)

if resolution:
    print(f"The resolution of the image is: {resolution[0]}x{resolution[1]}")
else:
    print("Failed to get the image resolution.")

```

### RENAME FILES

```python
import os

def rename_images_to_segments(folder_path):
    # List of supported image extensions
    supported_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp')

    # Counter for the new file names
    counter = 1

    # Iterate over all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file has a supported image extension
        if filename.lower().endswith(supported_extensions):
            # Create the new name in the form of 'segments_{number}.extension'
            ext = os.path.splitext(filename)[1]
            new_name = f"segment_{counter}{ext}"
            # Get the full paths for the current and new names
            current_file_path = os.path.join(folder_path, filename)
            new_file_path = os.path.join(folder_path, new_name)
            # Rename the file
            os.rename(current_file_path, new_file_path)
            print(f"Renamed: {filename} to {new_name}")
            # Increment the counter
            counter += 1

# Replace 'your_folder_path' with the path to your folder containing the images
your_folder_path = r"F:\ISRO_BAH\output_segments_20211228"
rename_images_to_segments(your_folder_path)
```
